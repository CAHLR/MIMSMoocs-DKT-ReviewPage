```python
import pandas as pd
import numpy as np
import json
```

```python

"""READ ORIGINAL FULL DATAFRAME PREINDEX FILE IF NO SAVED RESPONSE AND SKILL FILES"""
data_csv=pd.read_csv("code/delft15_phase2/pre_index_data.csv").drop('Unnamed: 0',axis=1)
event_dict=json.load(open('code/delft15_phase2/all_event_dict_delft15.json'))
unique_user_id =set(filter(lambda x: x is not None, data_csv['user'].tolist()))
len(unique_user_id)

data_csv.head()
```

```python

"""delete all non-problem events"""
data_csv=data_csv[data_csv["is_problem"]==1]
"""CONVERT TIME TO TIMESTEPS"""
def create_timesteps_truncated(group,limit=100):
#     print(group)
    group=group.sort_values('timestamp')
#     assign timesteps from 1..
    indices=list(range(1,len(group["timestamp"])+1))
    length=len(indices)
    if length>limit:
#         print(length)
#         set all other timesteps to 0
        indices[limit:]=[0]*(length-limit)
    group["timesteps"]=indices
    return group

df=data_csv.groupby("user").apply(lambda group : create_timesteps_truncated(group))
# df.to_csv("delft15_phase1/delft_15_problems_timsteps_100.csv", index=False)


"""GET SPECIFIED TIMESTEPS"""
##get rows where timestep is valid that is within 100 or specified limit
dfvalid=df[df['timesteps']!=0]
dfvalid.shape

"""CONVERT EVENTNAME TO TAGS """
dfvalid['unique_representation_of_event']=dfvalid['unique_representation_of_event'].replace(event_dict)
dfvalid.to_csv('code/delft15_phase1'+'delft_15_event_stream.csv',index=False)



"""GET EVENT/SKILL DATAFRAME"""
event_df= dfvalid.pivot(index='user', columns='timesteps', values='unique_representation_of_event')
event_df=event_df.fillna(0)
event_df=event_df.astype(int)
event_df.to_csv('code/delft15_phase1/'+'problemevent_df_delft_15.csv')

"""GET RESPONSE DATAFRAME"""
correctness={"correct":int(1),"incorrect":int(0)}
response_data= dfvalid.pivot(index='user', columns='timesteps', values='is_correct')
response_df=response_data.replace(correctness) 
response_df=response_df.fillna(0)
response_df=response_df.astype(int)
response_df.to_csv('code/delft15_phase1/'+'problemresponse_df_delft_15.csv')
```
